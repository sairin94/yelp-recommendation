```python
%%time
# prototype collaborative filtering
query = """SELECT user_id, reviews.business_id, reviews.stars, reviews.date from reviews
join businesses on reviews.business_id = businesses.business_id
where businesses.state = 'IN'
limit 300000
"""
df = pd.read_sql(query, conn)

# reviews could be used in content-based filtering and in collaborative filtering
df
```

```python
df.to_csv("user-bus-star.csv", index=False)
```

```python
# get a collaborative filtering algorithm/model
import implicit
from scipy.sparse import csr_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
user_enc = LabelEncoder()
user_enc.fit(df.user_id.values)
df["user_id_i"] = user_enc.transform(df.user_id)

bus_enc = LabelEncoder()
bus_enc.fit(df.business_id.values)
df["business_id_i"] = bus_enc.transform(df.business_id)

df = df.merge(bus[["business_id", "name"]], on="business_id")

df
```

```python
model = implicit.als.AlternatingLeastSquares()
```

```python
# get the most recent review for each user-business pair
df = df.sort_values("date", ascending=False).drop_duplicates(subset=["user_id", "business_id"])

data = csr_matrix((df.stars.values/5, (df.user_id_i.values, df.business_id_i.values)))
```

```python
df.shape
```

```python
model.fit(data)
```

```python
user_id_sample = 52175
rec_i, scores = model.recommend(user_id_sample, data[user_id_sample],filter_already_liked_items=True )
```
